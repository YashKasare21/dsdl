### Imputation
### Types of Imputation : Mean, Median , Fequency 

```python
df['Age']
```

```python
def meanImputation (dataset, column, mean):
    dataset[column + 'Mean'] = dataset[column].fillna(mean)
```

```python
mean =  df.Age.mean()
```

```python
print(f'Mean of the age is {mean}')
```

```python
meanImputation(dataset= df, column = 'Age', mean = mean)
```

# Now using SkLearn with Simple Imputer 

```python
from sklearn.impute import SimpleImputer
```

```python
imputeMean = SimpleImputer(strategy = "mean")
```

```python
imputeMean.fit(df2[['Age']])
```

```python
df2['ageMean'] = imputeMean.transform (df2[['Age']])
```

```python
df2.head(2)
```

```python
imputeMedian = SimpleImputer (strategy='median')
imputeMedian.fit (df2[['Age']])
df2['ageMedian'] = imputeMedian.transform(df2[['Age']])
df2.head()
```
